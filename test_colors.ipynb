```python
import cv2
import os
import json
import numpy as np
from skimage.color import rgb2lab, deltaE_ciede2000
```

```python
def display_color(color1, color2):
    height, width = 200, 200
    color_image1 = np.zeros((height, width, 3), np.uint8)
    color_image1[:, :] = color1

    color_image2 = np.zeros((height, width, 3), np.uint8)
    color_image2[:, :] = color2

    cv2.imshow("1", color_image1)
    cv2.imshow("2", color_image2)

    #euclidean distance between the 2 colors
    normalDistance = np.sqrt(3 * (color1[0] - color2[0])**2 + 4 * (color1[1] - color2[1])**2 + 2 * (color1[2] - color2[2])**2)
    maximumDistance = np.sqrt(3 * (255)**2 + 4 * (255)**2 + 2 * (255)**2)
    ratio = normalDistance / maximumDistance
    print(f"Normalized Euclidean Distance between color 1 and color 2 is {ratio}")

    euclideanDistance = np.linalg.norm(np.array(color1) - np.array(color2))
    maximumEuclideanDistance = np.linalg.norm(np.array([255, 255, 255]) - np.array([0, 0, 0]))
    ratio2 = euclideanDistance / maximumEuclideanDistance
    print(f"Euclidean Distance between color 1 and color 2 is {ratio2}")

    #convert colors to cielab and calculate the distance

    color1 = cv2.cvtColor(np.uint8([[color1]]), cv2.COLOR_BGR2Lab)[0][0]
    color2 = cv2.cvtColor(np.uint8([[color2]]), cv2.COLOR_BGR2Lab)[0][0]


    labDistance = np.linalg.norm(np.array(color1) - np.array(color2))
    #print(f"CIELAB Euclidean Distance between color 1 and color 2 is {labDistance}")

    color1 = (color1[0], color1[1], color1[2])
    color2 = (color2[0], color2[1], color2[2])

    diff = deltaE_ciede2000(color1, color2)
    #print(f"Delta E CIE2000 between color 1 and color 2 is {diff}")

    print("\n\n\n\n")

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def display_colorLab(color1, color2):
    height, width = 200, 200
    color_image1 = np.zeros((height, width, 3), np.uint8)
    color_image1[:, :] = color1

    color_image2 = np.zeros((height, width, 3), np.uint8)
    color_image2[:, :] = color2

    cv2.imshow("1", color_image1)
    cv2.imshow("2", color_image2)

    #euclidian distance between 2 colors
    normalDistance = np.linalg.norm(np.array(color1) - np.array(color2))
    print(f"Normal Euclidean Distance between color 1 and color 2 is {normalDistance}")

    color1 = cv2.cvtColor(np.uint8([[color1]]), cv2.COLOR_BGR2Lab)[0][0]
    color2 = cv2.cvtColor(np.uint8([[color2]]), cv2.COLOR_BGR2Lab)[0][0]

    color1 = (color1[0], color1[1], color1[2])
    color2 = (color2[0], color2[1], color2[2])

    diff = deltaE_ciede2000(color1, color2)
    print(f"Delta E CIE2000 between color 1 and color 2 is {diff}")

    print("\n\n\n\n")

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
